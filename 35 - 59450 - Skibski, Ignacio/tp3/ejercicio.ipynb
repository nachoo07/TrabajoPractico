## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def cargar_datos_ventas(lineas):
    datos_ventas = []
    
    for linea in lineas:
        fecha = linea[0:10].strip()         
        producto = linea[10:40].strip()     
        precio = int(linea[40:50].strip())  
        cantidad = int(linea[50:55].strip())  
        
        registro = {
            'fecha': fecha,
            'producto': producto,
            'precio': precio,
            'cantidad': cantidad
        }
        
        
        datos_ventas.append(registro)
    
    return datos_ventas


datos = cargar_datos_ventas(lineas)

def calcular_totales(datos):
    total_importe = 0
    total_cantidad = 0

    for dato in datos:
        
        importe = dato['precio'] * dato['cantidad']

        total_importe += importe
        total_cantidad += dato['cantidad']

    return total_importe, total_cantidad

total_importe, total_cantidad = calcular_totales(datos)

print(f"Las ventas fueron de ${total_importe:.2f} en {total_cantidad} unidades")
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def extraer_datos(lineas):
    datos_ventas = []
    
    for linea in lineas:
      
        producto = linea[10:40].strip()      
        precio = float(linea[40:50].strip()) 
        cantidad = int(linea[50:55].strip()) 
        
        registro = {
            'producto': producto,
            'precio': precio,
            'cantidad': cantidad
        }

        datos_ventas.append(registro)
    
    return datos_ventas

def calcular_precio_promedio_por_producto(ventas):
    sumas_precios = {}  
    conteo_productos = {}  

    for venta in ventas:
        producto = venta['producto']
        precio = venta['precio']

        if producto in sumas_precios:
            sumas_precios[producto] += precio
            conteo_productos[producto] += 1
        else:
            sumas_precios[producto] = precio
            conteo_productos[producto] = 1

    precio_promedio = {}

    for producto in sumas_precios:
        precio_promedio[producto] = sumas_precios[producto] / conteo_productos[producto]
    
    return precio_promedio

ventas = extraer_datos(lineas)

precio_promedio = calcular_precio_promedio_por_producto(ventas)

print("Precio promedio por producto:")
for producto, precio in precio_promedio.items():
    print(f"{producto}: ${precio:.2f}")
    
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def extraer_datos(lineas):
    datos_ventas = []
    
    for linea in lineas:
        producto = linea[10:40].strip()     
        cantidad = int(linea[50:55].strip()) 
        
        registro = {
            'producto': producto,
            'cantidad': cantidad
        }
        
        datos_ventas.append(registro)
    
    return datos_ventas

def listar_unidades_por_producto(ventas):
    total_por_producto = {}  

    for venta in ventas:
        producto = venta['producto']
        cantidad = venta['cantidad']
        
        if producto in total_por_producto:
            total_por_producto[producto] += cantidad
        else:
            total_por_producto[producto] = cantidad

    return total_por_producto

def obtener_tres_productos_mas_vendidos(total_por_producto):
    productos_ordenados = sorted(total_por_producto.items(), key=lambda item: item[1], reverse=True)
    
    return productos_ordenados[:3]

ventas = extraer_datos(lineas)

total_por_producto = listar_unidades_por_producto(ventas)

tres_mas_vendidos = obtener_tres_productos_mas_vendidos(total_por_producto)

print("Los 3 productos más vendidos son:")
for producto, cantidad in tres_mas_vendidos:
    print(f"{producto}: {cantidad} unidades")
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
from datetime import datetime

def extraer_datos_con_fecha(lineas):
    datos_ventas = []
    
    for linea in lineas:
        fecha_str = linea[0:10].strip()      
        producto = linea[10:40].strip()      
        cantidad = int(linea[50:55].strip()) 
        
        fecha = datetime.strptime(fecha_str, '%Y-%m-%d')
        
        registro = {
            'fecha': fecha,
            'producto': producto,
            'cantidad': cantidad
        }
        
        datos_ventas.append(registro)
    
    return datos_ventas

def agrupar_por_mes_y_producto(ventas):
    ventas_por_mes = {} 

    for venta in ventas:
        
        mes = venta['fecha'].strftime('%Y-%m')
        producto = venta['producto']
        cantidad = venta['cantidad']
        
        
        if mes not in ventas_por_mes:
            ventas_por_mes[mes] = {}
        
        if producto not in ventas_por_mes[mes]:
            ventas_por_mes[mes][producto] = 0
        
        ventas_por_mes[mes][producto] += cantidad

    return ventas_por_mes

def imprimir_ventas_por_mes(ventas_por_mes):
    for mes, productos in ventas_por_mes.items():
        print(f"Ventas del mes {mes}:")
        for producto, cantidad in productos.items():
            print(f"  {producto}: {cantidad} unidades")
        print() 

ventas = extraer_datos_con_fecha(lineas)

ventas_por_mes = agrupar_por_mes_y_producto(ventas)

imprimir_ventas_por_mes(ventas_por_mes)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def extraer_datos(lineas):
    datos_ventas = []
    
    for linea in lineas:
        producto = linea[10:40].strip()      
        precio = float(linea[40:50].strip()) 
        cantidad = int(linea[50:55].strip()) 
        
        registro = {
            'producto': producto,
            'precio': precio,
            'cantidad': cantidad
        }
        
        datos_ventas.append(registro)
    
    return datos_ventas

def calcular_totales_por_producto(ventas):
    productos_totales = {}  

    for venta in ventas:
        producto = venta['producto']
        precio = venta['precio']
        cantidad = venta['cantidad']
        importe = precio * cantidad
        
       
        if producto in productos_totales:
            productos_totales[producto]['cantidad_total'] += cantidad
            productos_totales[producto]['importe_total'] += importe
            productos_totales[producto]['suma_precios'] += precio
            productos_totales[producto]['conteo_precios'] += 1
        else:
         
            productos_totales[producto] = {
                'cantidad_total': cantidad,
                'importe_total': importe,
                'suma_precios': precio,
                'conteo_precios': 1
            }
    
  
    for producto in productos_totales:
        productos_totales[producto]['precio_promedio'] = (
            productos_totales[producto]['suma_precios'] / productos_totales[producto]['conteo_precios']
        )
    
    return productos_totales

def ordenar_productos_alfabeticamente(productos_totales):
   
    productos_ordenados = sorted(productos_totales.items(), key=lambda item: item[0])
    return productos_ordenados

def imprimir_listado_productos(productos_ordenados):
    print(f"{'Producto':<30} {'Precio Promedio':<15} {'Cantidad Vendida':<20} {'Importe Total':<15}")
    print("="*80)
    
    for producto, datos in productos_ordenados:
        print(f"{producto:<30} ${datos['precio_promedio']:<15.2f} {datos['cantidad_total']:<20} ${datos['importe_total']:<15.2f}")

     
ventas = extraer_datos(lineas)


productos_totales = calcular_totales_por_producto(ventas)


productos_ordenados = ordenar_productos_alfabeticamente(productos_totales)

imprimir_listado_productos(productos_ordenados)
````
